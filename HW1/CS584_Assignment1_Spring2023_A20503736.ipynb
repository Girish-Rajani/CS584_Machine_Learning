1.  Explain the differences between supervised, semi-surpervised, weakly-supervised and unsupervised learning.




**Answer:**

Supervised Learning - This is where our data comprises of inputs that have a corresponding target value. In this case, the training data is said to be labelled. When looking at supervised learning, the two most common tasks are regression and classification. Since the data is labelled, while training, the model can learn features and know what to expect when looking at test data since the inputs and outputs are labeled. 

Semi-Supervised Learning - Combination of supervised and unsupervised learning. This is where our dataset has a small amount of it that is labeled and most is unlabeled. We would train the model using the labeled data and once it is performing well, we perform prediction on the unlabeled data and label it using those prediction. This is especially helpful when we don't have a large labeled dataset.

Weakly-Supervised Learning - Instead of manually labeling examples, one can use weakly supervised learning to label the training data. Noisy and imprecise sources are used to create these weak labels which although are inexpensive and low-quality, still can help to create a strong model and increase the level of supervision.

Unsupervised Learning - This is where our data is not labeled. This means that we are given the input data and the model must try and figure out the output without the context. Patterns have to be manually identified by the model to make sense of the data without supervision. This makes the learning process more complicated when compared to supervised learning. Three common tasks of unsupervised learning are clustering, association, and dimensionality reduction.

2. Explain the difference between regression and classification.



**Answer:**


Classification - Classification problems deal with qualitative results, that is, belonging to a certain class label. Classification problems can typically be in the form of binary classification or multi-class classification. Such examples can include, married/not married, spam/not spam, etc. For example, classification can be used to predict whether an email is spam or not given certain features like email header, body, sender address, etc. The performance of a classification model is determined by the percentage of samples it has classified correctly.

In regression, we are able to perform prediction on continuous values. Such examples can include, age, income, house prices, etc. For example, regression can be used to predict house prices quantitatively given certain features like number of rooms, square feet, etc. The performance of a regression model is determined by the Mean Squared Error (MSE). The lower the MSE, the better the model performed.

3. Explain how to perform N-fold cross-validation.




**Answer:**

When our model has less data to work with and we want to split the dataset into training, and testing datasets, the model may give a noisy estimate. To improve this prediction, we can use N-fold cross-validation which divides the data into N subsets. Therefore, if N = 3 then the data will be split into 3 folds. The N-1 folds will be used for training and the final remaining subset will be used for testing. This process is repeated for N number of subsets. We need to be careful on the number of folds we select since increasing N will result in more models needing to be trained and more expensive computations. We will see low bias and high variance with increased N so we will need to find a right balance. A more detailed explanation of how to perform N-fold cross-validation:


1. Shuffle the dataset randomly
2. Split the datasets into N groups (folds)
3. For each unique fold:
    - Take that fold as the test data set
    - Take the remaining N-1 folds as the training data set
    - Fit the model on the training data and evaluate it on the test data
    - Record the model evaluation score and then remove the model
4. Using all of the model evaluation scores, average the results to summarize the model performance

5. Explain the standard steps to perform machine learning.




**Answer:**

1. Data Collection - 
We will need to collect data so that our model can learn and detect patterns. Depending on the quality of the data collected, the model will perform accordingly because if we use inaccurate/irrelevant data then this will be reflected in the performance of our model. Datasets that have already been pre-collected can be retrieved from sources such as Kaggle or UCI Repository.

2. Data Preparation - 
Preparing the data is crucial to ensure there is no bias within the data. For example, shuffling the dataset is important because if we get a pre-collected dataset that is in ascending order and do a train/test split, the data should be randomized because while training, the model may learn data that is within range A-M then is tested on ranges within N-Z, which will not generalize well and will struggle to adapt to unseen data. An even distribution is key when preparing the data. Once the dataset is randomized, cleaned and missing values are dealt with, the dataset should be split into training and testing data.

3. Choosing a Model - 
Depending on the task at hand, the appropriate model needs to be selected. A common example is, if we are predicting numerical outputs, the appropriate regression model must be selected and if we are predicting our outputs belonging to a category/label then the appropriate classification model must be selected.

4. Train the Model - 
This is where our model will learn from the data that we have collected and prepared. The model that we have selected/built will be trained by learning patterns from the training dataset using some initial model parameters in order to perform prediction on the test data. Through iteratively training our model and making adjustments, it will learn to perform prediction with less error.

5. Evaluate the Model - 
After the model has been trained, we will want to evaluate it's performance on new unseen data. This will tell us if overfitting has occurred and also how well the model generalizes. Based on the error in this stage, we can optimize the model accordingly.

6. Parameter Tuning - 
To try and increase the accuracy or minimize the loss of our model, we can perform parameter tuning. Such parameters may include number of epochs, learning rate, initial theta values, etc. 

7. Perform Prediction - 
Once we have a model that generalizes well to new unseen data, we can use the model in the real world and make predictions.

7. Write the objective function that can be used to determine the regression model parameters. How is this objective function will be used to find model parameters?



**Answer:**

The objective function in Machine Learning seeks to minimize the loss by determining the optimal model parameters. A common object function that can be used here is the Mean Squared Error (MSE). The MSE Function can be illustrated below:

$MSE(\boldsymbol{\theta}) = \frac{1}{n}\sum_{i=1}^n(\hat{y}^{(i)} - y^{(i)})^2$

The MSE function takes the difference between the prediction and the true target, squares it, then finds the average across all n training samples. The smaller the MSE, the lower the loss is, which means the closer the model is fitting the data. From the MSE loss, gradient descent can be used to find optimal model parameters that can reduce the MSE loss and make the model fit the data better.

**Question: How did you choose your lr and epoch number?**

**Answer:**

```python
# gradient descent to find the optimal fit
lr = 0.1
epochs = 10000
def gd(theta, X, y, lr, epochs):
    loss_history = []
    for epoch in range(epochs):
        n = len(X)
        y_hat = theta[0] + theta[1]*X
        error = np.subtract(y_hat,y)
        theta[0] = theta[0] - lr*(sum(error)/n)
        theta[1] = theta[1] - lr*(np.sum((error)*X)/n)
        loss = loss_function(theta, X, y)
        loss_history.append(loss)
    return theta,loss_history

theta, loss_history = gd(theta, X, y, lr, epochs)
print(theta)

y_hat = theta[0] + theta[1]*X

#Reason for 10,000 epochs and lr 0.1 is because if we choose a less number such as 1,000 epochs or lr of 0.01 then our prediction plot would not match the example plot shown in part V
#In order to get the right model fit that is required, 10,000 epochs and lr of 0.1 was necessary for training
```
